### try:
exceptとelseに分岐
### except:
エラーが発生した場合の処理
### else:
エラーが発生しなかった場合の処理
### finally:
最終的な処理

## except:

### except Exception as e:
汎用的なエラーに対応

### except etc as e:
その他様々なエラーに個別に対応するものがある。

# Webスクレイピングでのエラー処理

```python
import requests
```

```python
url = "https://al.www.hogehoge.com/data"
try: 
    response = requests.get(url)
except requests.exceptions.HTTPError as e:
    print(f"HTTPエラー: {e}")
except requests.exceptions.RequestException as e:
    print(f"リクエストエラー: {e}")
else:
    print("データの取得に成功しました")
finally:
    print("Webスクレイピング処理終了")
```
